# Hello World em OpenCV

O `OpenCV (Open Computer Vision)` é uma biblioteca utilizada com frequência no ramo de Visão Computacional em Python.
Ela apresenta uma sintaxe simples e intuitiva, junto com diversas funções úteis para leitura e manipulação de imagens.

## Leitura e exibição de uma imagem

Vamos fazer a leitura da imagem *cat*, que está na pasta de imagens.

```python
obj_img = cv2.imread('imagens/cat.png')
```

Agora vamos exibir a imagem.

Podemos fazer isso de duas formas:

1. Usando a função `cv2.imshow()`
2. Usando a plotagem do `matplotlib`

```python
# Lendo com cv2.imshow()

cv2.imshow("Cat", obj_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Lendo com matplotlib 

plt.imshow(obj_img)
plt.show()
```

## Cores em OpenCV

Como podemos ver no plot acima, a imagem ficou com as cores diferentes, puxando para um tom mais azulado.

Por coincidência, isso aconteceu apenas na exibição com matplotlib.

Isso aconteceu porque o OpenCV faz a leitura e trabalha com as imagens no padrão **(B, G, R)**, ou seja, ele trabalha com o oposto de **(R, G, B)**.

O matplotlib, por sua vez, não trabalha exatamente com B,G,R, e por conta disso, ele acaba lendo e exibindo a imagem como se fosse em RGB comum.

Resultado:
- As cores azul e vermelha estão trocadas. O que era mais avermelhado ficou com tons azuis, e o que era mais azulado ficou com tons de vermelho.

Para solucionar isso, precisamos manipular as cores da imagem de B,G,R para R,G,B. O OpenCV possui uma função que faz isso.

```python
obj_img = cv2.cvtColor(obj_img, cv2.COLOR_BGR2RGB)
plt.imshow(obj_img)
plt.show()
```
